```python
# importing necessary libraries

import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
# Reading the file!!
all_data = pd.read_csv(r'C:\Users\34661\Downloads\women_clothing_ecommerce_sales.csv')
all_data.head()
```

# Lets do Data Cleaning!!

```python
nan_data = all_data[all_data.isna().any(axis = 1)]
nan_data
```

```python
# Dropping all null values!!!

all_data.dropna(inplace = True)
all_data
```

```python
# Finding info about the database that we have!!
all_data.info()
```

```python
# Getting the dtype
all_data.dtypes
```

# Now lets Ponder on Some Analysis

# 1. Best Selling SKU Items By Color and Size

```python
# Code for getting it
best = all_data.groupby(['sku', 'color', 'size'])['quantity'].sum().reset_index()
best_selling = best.sort_values(by = 'quantity', ascending = False)
best_selling.head()
```

```python
 # let us plot the graph!!

# Getting the top 10 Selling By Size
grouped = best_selling.head(10)

# Setting 'set' as the index for plotting
grouped.set_index(['sku', 'size', 'color'], inplace = True)

# Plotting!!
grouped.plot(kind = 'bar', color = 'skyblue')

# Add labels and title

plt.xlabel('Size and Color')
plt.ylabel('Total Quantity Sold')
plt.title('Total Quantity Sold by Size')

# Show the plot
plt.show()
```

# Worst SKU Items By Color and By Size

```python
worst = all_data.groupby(['sku', 'color', 'size'])['quantity'].sum().reset_index()
worst_selling = worst.sort_values(by = 'quantity', ascending = True)
worst_selling.head(10)
```

# 2. What is the average order value? 

```python
all_data.head()
```

```python
# Finding average order value

revenue = all_data['revenue'].sum()
orders = all_data['order_id'].nunique()
average_order_value = revenue / orders
print('Average Order Value:', average_order_value)
```

# 3. What are the peak days with the highest revenue


```python
# Converting 'order date' to datetime
all_data['order_date'] = pd.to_datetime(all_data['order_date'])

# Extracting day, month, hour, minute, second
all_data['day'] = all_data['order_date'].dt.day
all_data['month'] = all_data['order_date'].dt.month
all_data['hour'] = all_data['order_date'].dt.hour
all_data['minute'] = all_data['order_date'].dt.minute
all_data['second'] = all_data['order_date'].dt.second

# Printing to check the output
all_data.head()
```

```python
days = all_data.groupby('day')['revenue'].count().reset_index()
best_days = days.sort_values(by = 'revenue', ascending = False)
best_days.head(10)
```

```python
# let us plot the graph!!

# Getting the top 10 Selling By Size
grouped = best_days.head(10)

# Setting 'set' as the index for plotting
grouped.set_index(['day'], inplace = True)

# Plotting!!
grouped.plot(kind = 'bar', color = 'skyblue')

# Add labels and title

plt.xlabel('Day')
plt.ylabel('Revenue')
plt.title('Peak Days That Bring High Revenue')

# Show the plot
plt.show()
```

# 4. Which peak months bring high revenue?

```python
all_data.head()
```

```python
months = all_data.groupby('month')['revenue'].count().reset_index()
best_months = months.sort_values(by = 'revenue', ascending = False)
best_months
```

```python
# let us plot the graph!!

# Getting the top 10 Selling By Size
grouped = best_months.head()

# Setting 'set' as the index for plotting
grouped.set_index(['month'], inplace = True)

# Plotting!!
grouped.plot(kind = 'bar', color = 'skyblue')

# Add labels and title

plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Peak Months That Bring High Revenue')

# Show the plot
plt.show()
```

```python
hours = all_data.groupby('hour')['revenue'].count().reset_index()
best_hours = hours.sort_values(by = 'revenue', ascending = False)
best_hours
```
